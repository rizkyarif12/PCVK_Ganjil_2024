```python
urls = ["https://i.pinimg.com/1200x/e5/02/e8/e502e8765b7488a7fcf11802a3fb4481.jpg"]

panjang_garis = 80

for url in urls:
    image = io.imread(url)  # Baca gambar dari URL
    image_resized = cv.resize(image, (0, 0), fx=0.5, fy=0.5)  # Resize gambar menjadi setengah ukuran

    # Konversi gambar ke format RGB
    image_rgb = cv.cvtColor(image_resized, cv.COLOR_BGR2RGB)

    # Dapatkan dimensi gambar
    tinggi, lebar, _ = image_rgb.shape

    # Salin gambar untuk manipulasi
    image_with_vertical_line = image_rgb.copy()
    image_with_diagonal_line = image_rgb.copy()

    # Membuat garis vertikal putih di tengah gambar
    cv.line(image_with_vertical_line, (int(lebar / 2), 0), (int(lebar / 2), tinggi - 1), (255, 255, 255), 1)

    # Membuat garis diagonal putih menyilang pada gambar
    if tinggi != lebar:
        cv.line(image_with_diagonal_line, (0, 0), (lebar - 1, tinggi - 1), (255, 255, 255), 1)
        cv.line(image_with_diagonal_line, (0, tinggi - 1), (lebar - 1, 0), (255, 255, 255), 1)
    else:
        cv.line(image_with_diagonal_line, (0, 0), (lebar - 1, tinggi - 1), (255, 255, 255), 1)
        cv.line(image_with_diagonal_line, (0, tinggi - 1), (lebar - 1, 0), (255, 255, 255), 1)

    # Gabungkan gambar dengan garis vertikal dan gambar dengan garis diagonal
    frame_1 = cv.hconcat((image_rgb, image_with_vertical_line))
    frame_2 = cv.hconcat((image_rgb, image_with_diagonal_line))

    # Tampilkan gambar
    cv2_imshow(frame_1)
    cv2_imshow(frame_2)
```
